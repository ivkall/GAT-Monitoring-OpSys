```python
import dgl
import torch
import torch.nn as nn
import torch.nn.functional as F
from dgl.nn import GATConv
import numpy as np
import pickle
from custom_gcn import DataPoint
import matplotlib.pyplot as plt

import dgl
import torch
import torch.nn as nn
import torch.nn.functional as F
from dgl.nn import GATConv
import numpy as np
import pickle
from custom_gcn import DataPoint
from sklearn.model_selection import train_test_split
from autoencode import Autoencoder
def standardize_data(data):
    """
    Standardizes the data to have a mean of 0 and standard deviation of 1.
    
    Parameters:
    data (np.array): The data to be standardized.

    Returns:
    np.array: The standardized data.
    """
    mean = np.mean(data, axis=0)
    std = np.std(data, axis=0)
    std[std == 0] = 1  # To avoid division by zero for constant features
    standardized_data = (data - mean) / std
    return standardized_data

def normalize_to_minus_one_and_one(data):
    min_vals = np.min(data, axis=0)
    max_vals = np.max(data, axis=0)
    ranges = max_vals - min_vals
    
    # Handle the case where the range is 0
    ranges[ranges == 0] = 1  # Prevent division by zero; these values will be set to 0 after normalization
    
    normalized_data = (data - min_vals) / ranges  # Scale to [0, 1]
    normalized_data = normalized_data * 2 - 1  # Scale to [-1, 1]
    
    # If the range is zero, set the normalized data to 0 for those features
    normalized_data[:, ranges == 0] = 0
    
    return normalized_data
# Define your Graph Attention Network model
class GAT(nn.Module):
    def __init__(self, in_dim, hidden_dim, out_dim, num_heads):
        super(GAT, self).__init__()
        self.layers = nn.ModuleList([
            GATConv(in_dim, hidden_dim, num_heads=num_heads),
            GATConv(hidden_dim * num_heads, out_dim, num_heads=1)
        ])

    def forward(self, g, features):
        x = features
        for layer in self.layers:
            x = layer(g, x).flatten(1)
            x = F.elu(x)
        return x

def load_data(filename):
    with open(filename, "rb") as f:
        data = pickle.load(f)
    return data

training_data = load_data("data1_1.pkl")

# Example adjacency matrix as a NumPy array
adjacency_matrix_np = np.array(training_data[0].adjacency_matrix)

######################################################

for i in range(len(adjacency_matrix_np)):
   
    adjacency_matrix_np[i][i] = 1
print(adjacency_matrix_np)


# Convert the adjacency matrix to COO format
coo = np.transpose(adjacency_matrix_np.nonzero())

# Extract source and destination nodes from the COO format
src_nodes = coo[:, 0]
dst_nodes = coo[:, 1]

# Convert the COO format to a DGL graph
graph = dgl.graph((src_nodes, dst_nodes))

# Example node features as a NumPy array
node_features_np = np.array(training_data[0].features)
node_features_np_normalized = normalize_to_minus_one_and_one(node_features_np)

# Convert the normalized node features to a PyTorch tensor
node_features_tensor = torch.tensor(node_features_np_normalized, dtype=torch.float32)
# Define your GAT model
in_dim = node_features_np.shape[1]  # Input feature dimension
hidden_dim = 64  # Hidden dimension
# out_dim = 128  # Output dimension
out_dim = node_features_np.shape[1]  # Output dimension should match the number of features per node
num_heads = 8  # Number of attention heads
gat_model = GAT(in_dim, hidden_dim, out_dim, num_heads)
# Define the output dimension to match the number of features per node




# Handling the case where std is 0 to avoid division by zero
node_features_np_normalized = np.nan_to_num(node_features_np_normalized)


# Handling the case where max_val equals min_val to avoid division by zero
node_features_tensor = torch.tensor(node_features_np_normalized, dtype=torch.float32)
# Convert node features to PyTorch tensor and cast to float32



# Define loss function and optimizer
criterion = nn.MSELoss()  # Mean Squared Error loss for unsupervised learning
optimizer = torch.optim.Adam(gat_model.parameters(), lr=0.001)

train_loss = []
# Training loop
num_epochs = 500
for epoch in range(num_epochs):
    logits = gat_model(graph, node_features_tensor)
    loss = criterion(logits, node_features_tensor)  # Reconstruction loss
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    print(f"Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item()}")
    train_loss.append(loss.item())
#__________________________________________________________________________________________________________________
#__________________________________________________________________________________________________________________
#__________________________________________________________________________________________________________________
#__________________________________________________________________________________________________________________
    
# Extract features from intermediate layers
def extract_features(gat_model, g, features):
    gat_model.eval()  # Set the model to evaluation mode
    with torch.no_grad():  # No need to compute gradients
        for layer in gat_model.layers:
            features = layer(g, features).flatten(1)
            features = F.elu(features)
    return features
training_data = load_data("data1_100.pkl")

aggregated_features_list = []

for data_point in training_data:
    # Process each graph's adjacency matrix
    adjacency_matrix_np = np.array(data_point.adjacency_matrix)
    np.fill_diagonal(adjacency_matrix_np, 1)  # Add self-loops

    # Convert the adjacency matrix to COO format and then to a DGL graph
    src_nodes, dst_nodes = np.nonzero(adjacency_matrix_np)
    graph = dgl.graph((src_nodes, dst_nodes))

    # Normalize and convert node features to tensor
    node_features_np = np.array(data_point.features)
   
    node_features_tensor = torch.tensor(node_features_np, dtype=torch.float32)

    # Use the trained GAT model to aggregate features for each graph
    aggregated_features = extract_features(gat_model, graph, node_features_tensor)
    aggregated_features_list.append(aggregated_features)
print(len( aggregated_features_list))

# Concatenate all aggregated features from all graphs
all_aggregated_features = torch.cat(aggregated_features_list, dim=0)

# Now, you have the aggregated features for all nodes in all graphs
# Next, you can scale these features
all_aggregated_features_scaled = standardize_data(all_aggregated_features.detach().numpy())

# Finally, split the aggregated and scaled features into training and validation sets
train_data, val_data = train_test_split(all_aggregated_features_scaled, test_size=0.2)
out_dim = node_features_np.shape[1]  # This is used as the out_dim for your GAT model.

# For the autoencoder:
input_dim = out_dim  # The input dimension for the autoencoder is the output dimension of your GAT.
encoding_dim = 32  # Example: set encoding_dim to 32. Adjust based on your needs.
# Initialize and train the autoencoder
autoencoder = Autoencoder(input_dim, encoding_dim)
train_data = torch.tensor(train_data, dtype=torch.float32).detach().numpy()
val_data = torch.tensor(val_data, dtype=torch.float32).detach().numpy()
history = autoencoder.train(train_data, val_data)

# Evaluate the model and visualize the reconstruction quality
reconstructed_data = autoencoder.evaluate(val_data)

# Select random samples to display their original and reconstructed values
num_samples_to_display = 5
indices = np.random.choice(range(len(val_data)), num_samples_to_display)

print("Comparing original and reconstructed values for random data points:")
for i, index in enumerate(indices):
    print(f"\nData Point {i + 1} (Index {index}):")
    
    print(list(zip(val_data[index][:10], reconstructed_data[index][:10])))

# Visualizing the reconstruction for selected samples
train_loss = history.history['loss']
val_loss = history.history['val_loss']
print("Training Loss: ", train_loss[-1])
print("Validation Loss: ", val_loss[-1])
```

```python
# Plot the training and validation loss
plt.figure(figsize=(10, 5))
plt.plot(train_loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
